```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv('tmdb_5000_movies.csv')
credits_df = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies_df.merge(credits_df,on='title')
```

```python
movies = movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
movies.dropna(axis=0,inplace=True)
```

```python
movies.iloc[0].genres
```

```python
#the above data is a list of dictionaries and i want to convert them into ['Action','Adventure','Fantasy','Science Fiction']
movies.shape
```

```python
import ast
def convert_to_list(obj):
  genres_list = []
  for i in ast.literal_eval(obj):
    genres_list.append(i['name'])
  return genres_list

movies['genres'] = movies['genres'].apply(convert_to_list)




```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert_to_list)
```

```python
movies.head(1)
```

```python
movies.iloc[0].cast
```

```python
import ast
def convert_to_list_cast(obj):
  counter_list = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:
      counter_list.append(i['name'])
      counter +=1
    else:
      break
  return counter_list

movies['cast'] = movies['cast'].apply(convert_to_list_cast)
```

```python
movies.head(1)
```

```python
movies.iloc[0].crew
```

```python
import ast
def convert_to_list_crew(obj):
  crew_item = []
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      crew_item.append(i['name'])
      break
  return crew_item

movies['crew'] = movies['crew'].apply(convert_to_list_crew)
```

```python
movies.head(1)
```

```python
movies.iloc[0].overview
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x ])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['cast']+movies['crew']+movies['keywords']
```

```python
movies.head()
```

```python
new_movies = movies[['id','title','tags']]
new_movies.head()
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  string = ' '.join(y)
  return string
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_movies[new_movies['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
